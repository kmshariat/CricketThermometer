### Finding the best fit line to the data

```python
#importing necesssary modules

import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = pd.read_csv("chirps.csv") #loading the dataset

#getting data from the csv file
temp = data.Temp_C 
N = data.Chirps_15s

#setting the graph level
plt.title('Data Visualization of Snowy Cricket')
plt.xlabel('Temperature')
plt.ylabel('No. of Chirps in 15 seconds')

plt.scatter(temp, N, marker='+')

#linear regressiong best fit line
a,b = np.polyfit(temp, N,1)
plt.plot(temp, a*temp+b, color='red')
plt.show()

print(f"Equation of the best fit line is: N = ({round(a,2)})t+({round(b,2)})")
```

### Finding the Co-relation Index

We are going to find out the Pearson correlation coefficient using numpy module. If the value of r> 0.8 then there is a strong positive co-relation between these two data.

```python
#using corrcoef function to find the corelation coefficient matrix of the data
r = np.corrcoef(temp,N)

#printing the r-index value
print(f"The value of r-index is: {r[0,1]}")
```

### Co-relation isn't Causation 

```python
#loading the dataset for Ice Cream Sale and Violent Crimes 
#the data ranges from January 1, 2021 to January 31, 2010

data = pd.read_csv('IceCreamCrimeData.csv')
revenue = data.Revenue
crime = data.Interpersviolence
```

```python
#plotting the data
plt.scatter(revenue,crime,marker=".")

#linear regressiong best fit line
a1,b1,c1= np.polyfit(revenue, crime,2)
plt.plot(revenue, a1*revenue**2+b1*revenue+c1, color='red')


#plotting the best fit line
#plt.tick_params(axis='both', left=False, top=False, right=False, bottom=False, labelleft=False, labeltop=False, labelright=False, labelbottom=True)
plt.xlabel('Revenue from Ice Cream Selling')
plt.ylabel('Interpersviolence')
plt.title('Co-relation between Ice Cream Sales and Crime Rates')
plt.show()
```

```python
#best fit line equation
print(f"Equation of the best fit curve is: r = ({round(a1,2)})t^2+({round(b1,2)})t+({round(c1,2)})")

#using corrcoef function to find the corelation coefficient matrix of the data
r = np.corrcoef(revenue,crime)

#printing the r-index value
print(f"The value of r-index is: {round(r[0,1],2)}")
```
